```python
data1 = 2

for i in range(1, 10):
    print(data1, 'X', i, '=',  data1 * i)
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(i, 'X', j, '=',  i * j)
```

```python
datas1 = list()

for i in range(2, 10):
    for j in range(1, 10):
        datas1.append(i * j)
        
datas1
```
